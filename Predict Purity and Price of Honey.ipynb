# Assumption 1: Linearity CheckÂ¶
1. Ho : There is no Linearity between Independent and Dependent variable

2. Ha : There is some Linearity between Independent and Dependent variable

```python
from scipy.stats import pearsonr
a=0.05
for i in l:
    p_stat, pvalue=pearsonr(df[i],df['Price'])
    print('::',i,'::')
    if pvalue<a:
        print('Reject Null Hypothesis. There is some Linearity')
    else:
        print('Failed to reject Null Hypothesis. There is no Linearity')
    print('*'*55)
```

```python
X = df.drop(columns=['Price'])
y=df['Price']
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
def calculate_vif(df, thresh=5.0):
    variables = list(range(df.shape[1]))
    dropped_columns = []  # List to store names of dropped columns
    dropped = True
    while dropped:
        dropped = False
        vif = [variance_inflation_factor(df.iloc[:, variables].values, ix) for ix in variables]
        
        max_vif = max(vif)
        if max_vif > thresh:
            maxloc = vif.index(max_vif)
            dropped_columns.append(df.columns[variables[maxloc]])  # Store name of dropped column
            df.drop(df.columns[variables[maxloc]], axis=1, inplace=True)
            del variables[maxloc]
            dropped = True
    
    return df, dropped_columns

# Calculate VIF and drop columns iteratively
X, dropped_columns = calculate_vif(X)

print("Columns after VIF selection:")
print(X.head())

print("\nDropped columns:")
print(dropped_columns)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=42, test_size=0.2)
```

```python
model.fit(X_train,y_train)
```

```python
y_pred_train = model.predict(X_train)
y_pred_test = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score

print('Model Accuracy Train :',r2_score(y_train,y_pred_train)*100,'%')

print('Model Accuracy Test  :',r2_score(y_test,y_pred_test)*100,'%')
```
